## Parte 1

```python
# Firma pregunta 1: a)
'''
input:
x       : (float)    variable a evaluar en f1.
t       : (float)    parámetro constante perteneciente a f1.


output:
eval    : (float) resultado de evaluar la función f1 con parámetros x, t.
'''
def f1(x, t):
    # Acá va su código
    x = float(x)
    t = float(t)
    a = (t+x*np.sin(x))
    b = np.sqrt(np.power(t,2)+1)
    eval = float(a + b)
    return eval


# Firma pregunta 1: b)
'''
input:
x       : (float)    variable a evaluar en f2.
t       : (float)    parámetro constante perteneciente a f2.


output:
eval    : (float) resultado de evaluar la función f2 con parámetros x, t.
'''
def f2(x, t):
    # Acá va su código
    x = float(x)
    t = float(t)
    a = x-float(np.power((x-t),5))
    b = t
    eval = a/b
    return eval


# Firma pregunta 1: c)
'''
input:
x       : (float)    variable a evaluar en f3.
t       : (float)    parámetro constante perteneciente a f3.


output:
eval    : (float) resultado de evaluar la función f3 con parámetros x, t.
'''
def f3(x, t):
    # Acá va su código
    x = float(x)
    t = float(t)
    a = (t+x*np.sin(x))
    b = np.sqrt(np.power(t,2)+1)
    c = float(np.power((x-t),5))
    d = (x-c)/t
    eval = float(a + b + d)
    return eval

# Firma pregunta 1: d)
'''
input:
x       : (float)    variable a evaluar en f1_fixed.
t       : (float)    parámetro constante perteneciente a f1_fixed.


output:
eval    : (float) resultado de evaluar la función f1_fixed con parámetros x, t.
'''
def f1_fixed(x, t):
    # Acá va su código
    x = float(x)
    t = float(t)
    a = t + np.sqrt(np.power(t,2)+1)
    b = x*np.sin(x) 
    eval = float(a + b)
    return eval

# Firma pregunta 1: e)
'''
input:
x       : (float)    variable a evaluar en f2_fixed.
t       : (float)    parámetro constante perteneciente a f2_fixed.


output:
eval    : (float) resultado de evaluar la función f2_fixed con parámetros x, t.
'''
def f2_fixed(x, t):
    # Acá va su código
    x = float(x)
    t = float(t)
    acum = np.power(t,4) - 5*np.power(t,3)*x
    acum += 10*np.power(t,2)*np.power(x,2)
    acum += -np.power(x,5)/t 
    acum += -10*t*np.power(x,3)
    acum += x/t
    acum += 5*np.power(x,4) 
    eval = float(acum)
    return eval


# Firma pregunta 1: f)
'''
input:
x       : (float)    variable a evaluar en f3_fixed.
t       : (float)    parámetro constante perteneciente a f3_fixed.


output:
eval    : (float) resultado de evaluar la función f3_fixed con parámetros x, t.
'''
def f3_fixed(x, t):
    # Acá va su código
    x = float(x)
    t = float(t)
    eval = float(f1_fixed(x, t) + f2_fixed(x, t))
    return eval
```

```python
# Firma pregunta 2: a)
'''
input:
x       : (float)    variable a evaluar en f2_diff.
t       : (float)    parámetro constante perteneciente a f2_diff.

output:
eval    : (float) resultado de evaluar la función f2_diff con parámetros x, t.
'''
def f2_diff(x, t):
    # Acá va su código
    x = float(x)
    t = float(t)
    acum = -5*np.power(t,3) + 20*np.power(t,2)*x
    acum += -30*t*np.power(x,2)
    acum += (-5*np.power(x,4)+1)/t
    acum += 20*np.power(x,3)
    eval = float(acum)
    return eval


# Firma pregunta 2: b)
'''
input:
f           : (callable) función a calcular la raíz usando método de Newton.
fp          : (callable) derivada de f.
x0          : (float)    initial guess para el método de newton.
t           : (float)    parámetro a mantener fijo en la evaluación de las funciones.
rel_error   : (float)    error relativo como criterio de parada (ver jupyters del curso).

output:
root        : (float)    raíz encontrada por el método de newton
'''
#################
# Implementación y modificación del método de Newton obtenido de 'https://github.com/tclaudioe/Scientific-Computing/blob/master/SC1v2/03_roots_of_1D_equations.ipynb'
#################
def newton_method(f, fp, x0, t, rel_error=1e-8):
    x0 = float(x0)
    maxNumberIterations = 100
    #Initialization of hybrid error and absolute
    hybrid_error = 100
    error_i = np.inf
    
    #Iteration counter
    i = 0
    while (hybrid_error > rel_error and hybrid_error < 1e12 and i<=maxNumberIterations):
        #Newton's iteration
        x1 = x0-f(x0,t)/fp(x0,t)
        
        #Checking if root was found
        if f(x1,t) == 0.0:
            hybrid_error = 0.0
            break
        
        #Computation of hybrid error
        hybrid_error = abs(x1-x0)/np.max([abs(x1),1e-12])
        
        #Computation of absolute error
        error_iminus1 = error_i
        error_i = abs(x1-x0)     
        
        #Updating solution
        x0 = x1
        #Increasing iteration counter
        i += 1
    
    root = float(x1)
    #Checking if solution was obtained
    if hybrid_error < rel_error: 
        return root
    elif i>=maxNumberIterations:
        return None
    else:
        return None

# Firma pregunta 2: c)
'''
input:
f           : (callable) función a calcular la raíz con metodo de la secante.
x0          : (float)    initial guess 1 para el método de la secante.
x1          : (float)    initial guess 2 para el método de la secante.
iters       : (int)      cantidad maxima de iteraciones como criterio de parada.
t           : (float)    parámetro a mantener fijo en la evaluación de las funciones.

output:
root    : (float) raíz encontrada por el método de newton
'''
def secant(f, x0, x1, t, iters=100):
    # Acá va su código
    x0 = float(x0)
    x1 = float(x1)
    for i in range(iters):
      if(f(x1,t)-f(x0,t) == 0):
        break
      x2 = x1-(f(x1,t)*(x1-x0))/(f(x1,t)-f(x0,t))
      x0 = x1
      x1 = x2
      root = float(x2)
    return root
```
